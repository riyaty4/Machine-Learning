# Matplotlib is a python library used to data visualization

```python
import numpy as np
```

```python
from matplotlib import pyplot as plt
```

```python
x = np.arange(100,10000)
```

```python
x
```

```python
y = 5 * x
```

```python
y
```

```python
plt.plot (x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.title('Line Plot')
plt.xlabel('X - Label')
plt.ylabel('Y - Label')
plt.show()
```

```python
plt.plot(x,y,color='g',linestyle='--',linewidth=5)
plt.title('Line Plot')
plt.xlabel('X - Label')
plt.ylabel('Y - Label')
plt.show()
```

```python
a = np.arange(10,100)
b1 = 2*a
b2 = 3*a
```

```python
plt.plot(a,b1,color='g',linestyle='--',linewidth=1)
plt.plot(a,b2,color='b',linestyle=':',linewidth=3)
plt.title('Line Plot')
plt.xlabel('X - Label')
plt.ylabel('Y - Label')
plt.show()
```
